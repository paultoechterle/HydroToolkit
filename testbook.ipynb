```python
import HydroToolkit as htk
import pandas as pd
```

```python
directory = 'HydroToolkit\data\ehyd_messstellen_7_qu' # search directory
csv_files = htk.list_files(directory)
path_lf, path_q, path_t = htk.filter_by_string(csv_files, '396317')

# Define paths and metadata for a Spring dataset
paths_spring = {
    'Q': path_q,
    'LF': path_lf,
    'TEMP': path_t
}

# Define paths and metadata for a River dataset
paths_river = {
    'Q': 'HydroToolkit\data\Q-WT-Tirol\QAlmdorf.csv',
    'TEMP': 'HydroToolkit\data\Q-WT-Tirol\WTAlmdorf.csv'
}

# Create Spring and River objects
spring_station = htk.Spring(paths_spring, path_lf, resolution='D')
river_station = htk.River(paths_river, 'HydroToolkit\data\Q-WT-Tirol\QAlmdorf.csv', resolution='D')
```

```python
timeslices = [
    [pd.Timestamp('2014-01-01'), pd.Timestamp('2014-03-18')],
    [pd.Timestamp('2016-11-28'), pd.Timestamp('2017-02-08')],
    ]
spring_station.plot_trockenfalllinien(timeslices)
```

```python
alpha = 0.0032

spring_station.print_aquifer_metrics(alpha)
```
